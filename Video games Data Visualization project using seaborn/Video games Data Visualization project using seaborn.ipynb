```python
platform_counts = df['Platform'].value_counts().head(10)

plt.figure(figsize=(12, 8))
bars = platform_counts.plot(kind='bar')
plt.title('Number of Games Released on Each Platform (Top 10)')
plt.xlabel('Platform')
plt.ylabel('Number of Games')
plt.xticks(rotation=45)

for bar in bars.patches:
    bars.annotate(format(bar.get_height(), '.0f'), 
                  (bar.get_x() + bar.get_width() / 2, bar.get_height()), 
                  ha='center', va='center', 
                  size=10, xytext=(0, 8), 
                  textcoords='offset points')

plt.show()
```

```python
platform_sales = df.groupby('Platform')['Global_Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(14, 8))
bars = platform_sales.plot(kind='bar')
plt.title('Top 10 Platforms by Total Global Sales')
plt.xlabel('Platform')
plt.ylabel('Total Global Sales (in millions)')
plt.xticks(rotation=45)

for bar in bars.patches:
    bars.annotate(format(bar.get_height(), '.2f'), 
                  (bar.get_x() + bar.get_width() / 2, bar.get_height()), 
                  ha='center', va='center', 
                  size=10, xytext=(0, 8), 
                  textcoords='offset points')

plt.show()
```

```python
plt.figure(figsize=(12, 6))
sales_over_years = df.groupby('Year')['Global_Sales'].sum()
plt.fill_between(sales_over_years.index, sales_over_years.values, color='skyblue', alpha=0.4)
plt.plot(sales_over_years.index, sales_over_years.values, marker='o', color='skyblue')
plt.title('Global Sales Over Years')
plt.xlabel('Year')
plt.ylabel('Global Sales (millions)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 8))
genre_counts = df['Genre'].value_counts()
plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Genre Distribution')
plt.axis('equal')
plt.show()
```

```python
publisher_counts = df['Publisher'].value_counts()

top_publishers = publisher_counts.head(15)
other_publishers = pd.Series(publisher_counts[15:].sum(), index=['Others'])

publisher_counts = pd.concat([top_publishers, other_publishers])

plt.figure(figsize=(12, 12))
publisher_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Games by Top 15 Publishers and Others')
plt.ylabel('')  # Hide the y-label for better appearance
plt.show()
```

```python
df = df.dropna(subset=['Year'])

df['Year'] = df['Year'].astype(int)

def get_top_platforms_per_year(group):
    return group.groupby('Platform').size().nlargest(5)

top_platforms_per_year = df.groupby('Year').apply(get_top_platforms_per_year).reset_index(name='Count')
pivot_table = top_platforms_per_year.pivot(index='Year', columns='Platform', values='Count').fillna(0)

plt.figure(figsize=(14, 8))
pivot_table.plot(kind='bar', stacked=True, colormap='tab20', figsize=(14, 8))
plt.title('Number of Games Released on Top 5 Platforms for Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Games')
plt.legend(title='Platform', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(12, 8))
sns.boxplot(x='Platform', y='Global_Sales', data=df)
plt.title('Global Sales Distribution by Platform')
plt.ylabel('Global Sales (millions)')
plt.xlabel('Platform')
plt.xticks(rotation=45)
plt.show()
```

```python
regional_sales = df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()


plt.figure(figsize=(10, 6))
regional_sales.plot(kind='bar', stacked=True)
plt.title('Regional Sales Contribution')
plt.ylabel('Sales (millions)')
plt.xlabel('Region')
plt.xticks(rotation=0)
plt.show()
```

```python
df = df.dropna(subset=['Year'])

df['Year'] = df['Year'].astype(int)

def get_top_publishers(sales_column):
    return df.groupby('Publisher')[sales_column].sum().nlargest(5)

top_publishers_global = get_top_publishers('Global_Sales')
top_publishers_na = get_top_publishers('NA_Sales')
top_publishers_jp = get_top_publishers('JP_Sales')
top_publishers_eu = get_top_publishers('EU_Sales')

fig, axs = plt.subplots(2, 2, figsize=(16, 12))

axs[0, 0].bar(top_publishers_global.index, top_publishers_global.values, color='tab:blue')
axs[0, 0].set_title('Top 5 Publishers by Global Sales')
axs[0, 0].set_xlabel('Publisher')
axs[0, 0].set_ylabel('Global Sales (in millions)')
axs[0, 0].tick_params(axis='x', rotation=45)
for i, v in enumerate(top_publishers_global.values):
    axs[0, 0].text(i, v + 1, f'{v:.2f}', ha='center')

axs[0, 1].bar(top_publishers_na.index, top_publishers_na.values, color='tab:orange')
axs[0, 1].set_title('Top 5 Publishers by NA Sales')
axs[0, 1].set_xlabel('Publisher')
axs[0, 1].set_ylabel('NA Sales (in millions)')
axs[0, 1].tick_params(axis='x', rotation=45)
for i, v in enumerate(top_publishers_na.values):
    axs[0, 1].text(i, v + 0.5, f'{v:.2f}', ha='center')

axs[1, 0].bar(top_publishers_jp.index, top_publishers_jp.values, color='tab:green')
axs[1, 0].set_title('Top 5 Publishers by JP Sales')
axs[1, 0].set_xlabel('Publisher')
axs[1, 0].set_ylabel('JP Sales (in millions)')
axs[1, 0].tick_params(axis='x', rotation=45)
for i, v in enumerate(top_publishers_jp.values):
    axs[1, 0].text(i, v + 0.1, f'{v:.2f}', ha='center')

axs[1, 1].bar(top_publishers_eu.index, top_publishers_eu.values, color='tab:red')
axs[1, 1].set_title('Top 5 Publishers by EU Sales')
axs[1, 1].set_xlabel('Publisher')
axs[1, 1].set_ylabel('EU Sales (in millions)')
axs[1, 1].tick_params(axis='x', rotation=45)
for i, v in enumerate(top_publishers_eu.values):
    axs[1, 1].text(i, v + 0.5, f'{v:.2f}', ha='center')

plt.tight_layout()
plt.show()
```

```python
top_genres_global = df.groupby('Genre')['Global_Sales'].sum().nlargest(10)

top_genres_na = df.groupby('Genre')['NA_Sales'].sum().nlargest(5)
top_genres_eu = df.groupby('Genre')['EU_Sales'].sum().nlargest(5)
top_genres_jp = df.groupby('Genre')['JP_Sales'].sum().nlargest(5)

fig, axs = plt.subplots(2, 2, figsize=(16, 12))

axs[0, 0].bar(top_genres_global.index, top_genres_global.values, color='tab:blue')
axs[0, 0].set_title('Top 10 Genres by Global Sales')
axs[0, 0].set_xlabel('Genre')
axs[0, 0].set_ylabel('Global Sales (in millions)')
axs[0, 0].tick_params(axis='x', rotation=45)
for i, v in enumerate(top_genres_global.values):
    axs[0, 0].text(i, v + 1, f'{v:.2f}', ha='center')

axs[0, 1].bar(top_genres_na.index, top_genres_na.values, color='tab:orange')
axs[0, 1].set_title('Top 5 Genres by NA Sales')
axs[0, 1].set_xlabel('Genre')
axs[0, 1].set_ylabel('NA Sales (in millions)')
axs[0, 1].tick_params(axis='x', rotation=45)
for i, v in enumerate(top_genres_na.values):
    axs[0, 1].text(i, v + 0.5, f'{v:.2f}', ha='center')

axs[1, 0].bar(top_genres_eu.index, top_genres_eu.values, color='tab:green')
axs[1, 0].set_title('Top 5 Genres by EU Sales')
axs[1, 0].set_xlabel('Genre')
axs[1, 0].set_ylabel('EU Sales (in millions)')
axs[1, 0].tick_params(axis='x', rotation=45)
for i, v in enumerate(top_genres_eu.values):
    axs[1, 0].text(i, v + 0.5, f'{v:.2f}', ha='center')

axs[1, 1].bar(top_genres_jp.index, top_genres_jp.values, color='tab:red')
axs[1, 1].set_title('Top 5 Genres by JP Sales')
axs[1, 1].set_xlabel('Genre')
axs[1, 1].set_ylabel('JP Sales (in millions)')
axs[1, 1].tick_params(axis='x', rotation=45)
for i, v in enumerate(top_genres_jp.values):
    axs[1, 1].text(i, v + 0.1, f'{v:.2f}', ha='center')

plt.tight_layout()
plt.show()
```
